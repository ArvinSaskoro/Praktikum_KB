```python
#menampilkan info dataset
import pandas as pd
import numpy as np

from prettytable import PrettyTable

df = pd.read_csv("drug_deaths.csv")

df.info()
```

>> Kodingan untuk poin 1 dan 2

```python
info_tabel = {
    "Jumah Record"                 : df.shape[0],
    "Nama Atribut"                 : list(df.columns.values),
    "Jumlhah Atribut Tipe Objek"   : list(df.select_dtypes(["object"]).columns),
    "Jumlah Atribut Tipe Angka"    : list(df.select_dtypes(["int64", "float64"]).columns),
    "Jumlah Atribut Null"          : list(df.loc[:, df.isnull().any()].columns),
}

keys = list(info_tabel.keys())

info_tabelTable = PrettyTable()
info_tabelTable.field_names = ["Info", "Atribut", "Jumlah"]

info_tabelTable.add_row([keys[0], "-", info_tabel[keys[0]]])

for i in range(1, 5):
    info_tabelTable.add_row([keys[i], ", \n".join(info_tabel[keys[i]]), len(info_tabel[keys[i]])])
    
print(info_tabelTable)
```
