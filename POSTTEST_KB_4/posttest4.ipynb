### DATA CLEANING
### Menangani Nilai NULL

```python
df = pd.read_csv("heart_2020_cleaned.csv")
df.isna().sum()
```

### Subtitusi Nilai NULL

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].median())

df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(strategy="mean")
imputer_modus = SimpleImputer(strategy="most_frequent")

df['Smoking'] = imputer_modus.fit_transform(df[['Smoking']])
df['AlcoholDrinking'] = imputer_modus.fit_transform(df[['AlcoholDrinking']])
df['Stroke'] = imputer_modus.fit_transform(df[['Stroke']])
df['PhysicalHealth'] = imputer_modus.fit_transform(df[['PhysicalHealth']])
df['MentalHealth'] = imputer_mean.fit_transform(df[['MentalHealth']])
df['DiffWalking'] = imputer_modus.fit_transform(df[['DiffWalking']])
df['Sex'] = imputer_modus.fit_transform(df[['Sex']])
df['AgeCategory'] = imputer_modus.fit_transform(df[['AgeCategory']])
df['Race'] = imputer_modus.fit_transform(df[['Race']])
df['Diabetic'] = imputer_modus.fit_transform(df[['Diabetic']])
df['PhysicalActivity'] = imputer_modus.fit_transform(df[['PhysicalActivity']])
df['GenHealth'] = imputer_modus.fit_transform(df[['GenHealth']])
df['SleepTime'] = imputer_modus.fit_transform(df[['SleepTime']])
df['Asthma'] = imputer_modus.fit_transform(df[['Asthma']])
df['KidneyDisease'] = imputer_modus.fit_transform(df[['KidneyDisease']])
df['SkinCancer'] = imputer_modus.fit_transform(df[['SkinCancer']])

df.isna().sum()
```
